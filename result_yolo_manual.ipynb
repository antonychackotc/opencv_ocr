```python
from ultralytics import YOLO
import cv2

# Load the model
model = YOLO('runs/detect/train/weights/best.pt')

# Run prediction
results = model('download.jpg')

# Loop through results and display using OpenCV
for r in results:
    im_array = r.plot()  # Plot boxes on image
    cv2.imshow("YOLO Detection", im_array)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
